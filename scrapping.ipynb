```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import time
import pandas as pd 
from bs4 import BeautifulSoup
```

```python
chromedriver_path = "chromedriver.exe"
```

```python
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service)
```

```python
URL = "http://localhost:8080"
```

```python
driver.get(URL)
time.sleep(3)
```

```python
html_source = driver.page_source
soup = BeautifulSoup(html_source, 'html.parser')
tarjetas = soup.find_all('div', class_='card-body')
datos_extraidos = []
for tarjeta in tarjetas:
    nombre_tag = tarjeta.find('h3')
    nombre = nombre_tag.text.strip() if nombre_tag else "Sin nombre"
    parrafos = tarjeta.find_all('p')
    horario = parrafos[0].text.replace("Hoy:", "").replace("🕒", "").strip() if len(parrafos) > 0 else "N/A"
    direccion = parrafos[1].text.replace("Dirección:", "").strip() if len(parrafos) > 1 else "N/A"
    calificacion = parrafos[2].text.replace("Calificación:", "").replace("⭐", "").strip() if len(parrafos) > 2 else "N/A"
    usuarios = parrafos[3].text.replace("Usuarios:", "").replace("👥", "").strip() if len(parrafos) > 3 else "0"
    datos_extraidos.append({
        'Nombre': nombre,
        'Horario': horario,
        'Dirección': direccion,
        'Calificación': calificacion,
        'Usuarios': usuarios
    })

print({len(datos_extraidos)})
```

```python
datos_extraidos
```

```python
df = pd.DataFrame(datos_extraidos)
df
```

```python
csv_generado = df.to_csv(index=False, encoding='utf-8-sig')
print("Vista previa del CSV generado:\n")
print(csv_generado[:500])
```
